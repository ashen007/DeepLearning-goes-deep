```python
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
vgg_11 = pd.read_pickle('../../images/computer vision/natural scene/original dataset training 96/vgg11_depth_comp.pkl')
vgg_13 = pd.read_pickle('../../images/computer vision/natural scene/original dataset training 96/vgg13_depth_comp.pkl')
vgg_16 = pd.read_pickle('../../images/computer vision/natural scene/original dataset training 96/vgg16_depth_comp.pkl')
vgg_19 = pd.read_pickle(
    '../../images/computer vision/natural scene/original dataset training 96/vgg19_compare_depth.pkl')
```

```python
vgg_11_aug = pd.read_pickle('../../images/computer vision/natural scene/augmented dataset 96/vgg11_depth_comp.pkl')
vgg_13_aug = pd.read_pickle('../../images/computer vision/natural scene/augmented dataset 96/vgg13_depth_comp.pkl')
vgg_16_aug = pd.read_pickle('../../images/computer vision/natural scene/augmented dataset 96/vgg16_depth_comp.pkl')
vgg_19_aug = pd.read_pickle('../../images/computer vision/natural scene/augmented dataset 96/vgg19_depth_comp.pkl')[:50]
```

```python
plt.figure(figsize=[12, 6], dpi=300)

sns.lineplot(x=list(range(vgg_11.shape[0])), y=vgg_11['val_accuracy'], color='#9BCAF2', label='vgg11')
sns.lineplot(x=list(range(vgg_13.shape[0])), y=vgg_13['val_accuracy'], color='#A69B03', label='vgg13')
sns.lineplot(x=list(range(vgg_16.shape[0])), y=vgg_16['val_accuracy'], color='#736B02', label='vgg16')
sns.lineplot(x=list(range(vgg_19.shape[0])), y=vgg_19['val_accuracy'], color='#594B02', label='vgg19')

plt.xlabel('epochs')
plt.ylabel('accuracy')

plt.savefig('../../images/computer vision/compare_depth.png')
plt.show()
```

```python
plt.figure(figsize=[12, 6], dpi=300)

sns.lineplot(x=list(range(vgg_11.shape[0])), y=vgg_11['val_accuracy'], color='#9BCAF2', label='vgg11')
plt.axvline(x=np.argmin(vgg_11['val_loss']), color='#9BCAF2', linestyle='--')
sns.lineplot(x=list(range(vgg_13.shape[0])), y=vgg_13['val_accuracy'], color='#A69B03', label='vgg13')
plt.axvline(x=np.argmin(vgg_13['val_loss']), color='#A69B03', linestyle='--')
sns.lineplot(x=list(range(vgg_16.shape[0])), y=vgg_16['val_accuracy'], color='#736B02', label='vgg16')
plt.axvline(x=np.argmin(vgg_16['val_loss']), color='#736B02', linestyle='--')
sns.lineplot(x=list(range(vgg_16.shape[0])), y=vgg_19['val_accuracy'], color='#594B02', label='vgg19')
plt.axvline(x=np.argmin(vgg_19['val_loss']), color='#594B02', linestyle='--')

plt.xlabel('epochs')
plt.ylabel('accuracy')

plt.savefig('../../images/computer vision/compare_depth_1.png')
plt.show()
```

```python
plt.figure(figsize=[12, 6], dpi=300)

sns.lineplot(x=list(range(vgg_11_aug.shape[0])), y=vgg_11_aug['val_accuracy'], color='#9BCAF2', label='vgg11')
sns.lineplot(x=list(range(vgg_13_aug.shape[0])), y=vgg_13_aug['val_accuracy'], color='#A69B03', label='vgg13')
sns.lineplot(x=list(range(vgg_16_aug.shape[0])), y=vgg_16_aug['val_accuracy'], color='#736B02', label='vgg16')
sns.lineplot(x=list(range(vgg_16_aug.shape[0])), y=vgg_19_aug['val_accuracy'][:vgg_16_aug.shape[0]],
             color='#594B02', label='vgg19')

plt.xlabel('epochs')
plt.ylabel('accuracy')

plt.savefig('../../images/computer vision/compare_depth_aug.png')
plt.show()
```

```python
plt.figure(figsize=[12, 6], dpi=300)

sns.lineplot(x=list(range(vgg_11_aug.shape[0])), y=vgg_11_aug['val_accuracy'], color='#9BCAF2', label='vgg11')
plt.axvline(x=np.argmin(vgg_11_aug['val_loss']), color='#9BCAF2', linestyle='--')
sns.lineplot(x=list(range(vgg_13_aug.shape[0])), y=vgg_13_aug['val_accuracy'], color='#A69B03', label='vgg13')
plt.axvline(x=np.argmin(vgg_13_aug['val_loss']), color='#A69B03', linestyle='--')
sns.lineplot(x=list(range(vgg_16_aug.shape[0])), y=vgg_16_aug['val_accuracy'], color='#736B02', label='vgg16')
plt.axvline(x=np.argmin(vgg_16_aug['val_loss']), color='#736B02', linestyle='--')
sns.lineplot(x=list(range(vgg_16_aug.shape[0])), y=vgg_19_aug['val_accuracy'][:vgg_16_aug.shape[0]],
             color='#594B02', label='vgg19')
plt.axvline(x=np.argmin(vgg_19_aug['val_loss']), color='#594B02', linestyle='--')

plt.xlabel('epochs')
plt.ylabel('accuracy')

plt.savefig('../../images/computer vision/compare_depth_aug_1.png')
plt.show()
```

```python
figure, axes = plt.subplots(nrows=2, ncols=2, figsize=[12, 12], dpi=300)
axes = axes.ravel()
epochs = 50
networks = ['vgg_11', 'vgg_13', 'vgg_16', 'vgg_19']
training_losses = [vgg_11, vgg_13, vgg_16, vgg_19]
training_losses_aug = [vgg_11_aug, vgg_13_aug, vgg_16_aug, vgg_19_aug]

for i in range(4):
    sns.lineplot(x=list(range(epochs)),
                 y=training_losses[i]['loss'],
                 marker='o',
                 label='training',
                 ax=axes[i])
    sns.lineplot(x=list(range(epochs)),
                 y=training_losses[i]['val_loss'],
                 marker='o',
                 label='validation',
                 ax=axes[i])

    axes[i].axvline(x=np.argmin(training_losses[i]['val_loss']), color='#F21905', linestyle='--')
    axes[i].set_title(networks[i])
    axes[i].set_xlabel('epoch')
    axes[i].set_ylabel('loss')

plt.savefig('../../images/computer vision/losses.png')
plt.show()
```

```python
figure, axes = plt.subplots(nrows=2, ncols=2, figsize=[12, 12], dpi=300)
axes = axes.ravel()

for i in range(4):
    sns.lineplot(x=list(range(epochs)),
                 y=training_losses_aug[i]['loss'],
                 marker='o',
                 label='training',
                 ax=axes[i])

    sns.lineplot(x=list(range(epochs)),
                 y=training_losses_aug[i]['val_loss'],
                 marker='o',
                 label='validation',
                 ax=axes[i])

    axes[i].axvline(x=np.argmin(training_losses_aug[i]['val_loss']), color='#F21905', linestyle='--')
    axes[i].set_title(networks[i])
    axes[i].set_xlabel('epoch')
    axes[i].set_ylabel('loss')

plt.savefig('../../images/computer vision/losses_aug.png')
plt.show()
```

```python
with_lrn = pd.read_pickle('../../images/computer vision/natural scene/original dataset training 150/vgg11_lrn.pkl')
without_lrn = pd.read_pickle('../../images/computer vision/natural scene/original dataset training 150/vgg11.pkl')

figure, axes = plt.subplots(nrows=1, ncols=2, figsize=[12, 6], dpi=300)

sns.lineplot(x=list(range(with_lrn.shape[0])), y=with_lrn['val_accuracy'], color='#9BCAF2', label='vgg11', ax=axes[0])
sns.lineplot(x=list(range(without_lrn.shape[0])), y=without_lrn['val_accuracy'], color='#A69B03', label='vgg11 LRN',
             ax=axes[0])
axes[0].axvline(x=np.argmin(with_lrn['val_loss']), color='#9BCAF2', linestyle='--')
axes[0].axvline(x=np.argmin(without_lrn['val_loss']), color='#A69B03', linestyle='--')

sns.lineplot(x=list(range(with_lrn.shape[0])), y=with_lrn['val_loss'], color='#9BCAF2', label='vgg11', ax=axes[1])
sns.lineplot(x=list(range(without_lrn.shape[0])), y=without_lrn['val_loss'], color='#A69B03', label='vgg11 LRN',
             ax=axes[1])
axes[1].axvline(x=np.argmin(with_lrn['val_loss']), color='#9BCAF2', linestyle='--')
axes[1].axvline(x=np.argmin(without_lrn['val_loss']), color='#A69B03', linestyle='--')

axes[0].set_xlabel('epochs')
axes[1].set_xlabel('epochs')
axes[0].set_ylabel('accuracy')
axes[1].set_ylabel('loss')

plt.savefig('../../images/computer vision/compare_lrn.png')
plt.show()
```
