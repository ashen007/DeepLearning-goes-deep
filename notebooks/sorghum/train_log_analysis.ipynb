```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
log_history = pd.read_pickle('submission/9/history.pkl')
log_history
```

```python
plt.figure(figsize=[12,6], dpi=300)
sns.lineplot(x=range(log_history.shape[0]),y=np.log(log_history.loss),label='loss')
sns.lineplot(x=range(log_history.shape[0]),y=np.log(log_history.val_loss),label='val loss')
plt.show()
```

```python
plt.figure(figsize=[12,6], dpi=300)
sns.lineplot(x=range(log_history.shape[0]),y=log_history.accuracy,label='accuracy')
sns.lineplot(x=range(log_history.shape[0]),y=log_history.val_accuracy,label='val accuracy')
plt.show()
```

```python
plt.figure(figsize=[12,6], dpi=300)
sns.lineplot(x=range(log_history.shape[0]),y=np.log(log_history.lr),label='learning rate')
plt.show()
```
