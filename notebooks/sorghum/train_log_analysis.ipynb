```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from matplotlib import pyplot as plt
```

```python
log_history = pd.read_pickle('submission/9/history.pkl')
log_history_2i = pd.read_pickle('submission/9/history-2nd-iter.pkl')
log_history_1 = pd.read_pickle('submission/10/history.pkl')
log_history_1_2nd = pd.read_pickle('submission/10/history-2nd-iter.pkl')
log_history_1_3rd = pd.read_pickle('submission/10/history-3rd-iter.pkl')
```

```python
plt.figure(figsize=[12, 6], dpi=300)
sns.lineplot(x=range(log_history.shape[0]), y=np.log(log_history.loss), label='loss', color='#590242')
sns.lineplot(x=range(log_history_2i.shape[0], log_history_2i.shape[0] * 2), y=np.log(log_history_2i.loss),
             label='loss 2nd iter', color='#590242', linestyle='--')
sns.lineplot(x=range(log_history.shape[0]), y=np.log(log_history.val_loss), label='val loss', color='#F2522E')
sns.lineplot(x=range(log_history_2i.shape[0], log_history_2i.shape[0] * 2), y=np.log(log_history_2i.val_loss),
             label='val loss 2nd iter', color='#F2522E', linestyle='--')
plt.show()
```

```python
plt.figure(figsize=[12, 6], dpi=300)
sns.lineplot(x=range(log_history.shape[0]), y=log_history.accuracy, label='accuracy', color='#590242')
sns.lineplot(x=range(log_history_2i.shape[0], log_history_2i.shape[0] * 2), y=log_history_2i.accuracy,
             label='loss 2nd iter', color='#590242', linestyle='--')
sns.lineplot(x=range(log_history.shape[0]), y=log_history.val_accuracy, label='val accuracy', color='#F2522E')
sns.lineplot(x=range(log_history_2i.shape[0], log_history_2i.shape[0] * 2), y=log_history_2i.val_accuracy,
             label='val loss 2nd iter', color='#F2522E', linestyle='--')
plt.show()
```

```python
plt.figure(figsize=[12, 6], dpi=300)
sns.lineplot(x=range(log_history.shape[0]), y=log_history.lr, label='learning rate', color='#F28705')
sns.lineplot(x=range(log_history_2i.shape[0]), y=log_history_2i.lr, label='learning rate', color='#347355',
             linestyle='--')
plt.show()
```

```python
fig = make_subplots(
    rows=2, cols=2,
    column_widths=[0.6, 0.4],
    row_heights=[0.4, 0.6],
    specs=[[{"type": "scatter", "rowspan": 2}, {"type": "scatter"}],
           [None, {"type": "scatter"}]])

fig.add_trace(go.Scatter(x=list(range(log_history_1.shape[0])), y=log_history_1.lr,
                         mode='lines',
                         name='learning rate',
                         marker=dict(color="#F28705")),
              row=1, col=1)

fig.add_trace(go.Scatter(x=list(range(log_history_1_2nd.shape[0])), y=log_history_1_2nd.lr,
                         mode='lines',
                         name='learning rate 2nd iteration',
                         line=dict(color="#347355", dash='dash')),
              row=1, col=2)

fig.add_trace(go.Scatter(x=list(range(log_history_1_3rd.shape[0])), y=log_history_1_3rd.lr,
                         mode='lines',
                         name='learning rate 3rd iteration',
                         line=dict(color="#8C0303", dash='dot')),
              row=2, col=2)

fig.update_layout(height=600,
                  width=1300,
                  plot_bgcolor='white')

fig.update_xaxes(showline=True, linecolor='black', gridcolor='#D9D9D9')
fig.update_yaxes(showline=True, linecolor='black', gridcolor='#D9D9D9')

fig.show()
```
