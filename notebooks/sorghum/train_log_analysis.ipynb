```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
log_history = pd.read_pickle('submission/9/history.pkl')
log_history_2i = pd.read_pickle('submission/9/history-2nd-iter.pkl')
log_history_1 = pd.read_pickle('submission/10/history.pkl')
```

```python
plt.figure(figsize=[12,6], dpi=300)
sns.lineplot(x=range(log_history.shape[0]),y=np.log(log_history.loss),label='loss', color='#590242')
sns.lineplot(x=range(log_history_2i.shape[0],log_history_2i.shape[0]*2),y=np.log(log_history_2i.loss),label='loss 2nd iter', color='#590242', linestyle='--')
sns.lineplot(x=range(log_history.shape[0]),y=np.log(log_history.val_loss),label='val loss', color='#F2522E')
sns.lineplot(x=range(log_history_2i.shape[0],log_history_2i.shape[0]*2),y=np.log(log_history_2i.val_loss),label='val loss 2nd iter', color='#F2522E', linestyle='--')
plt.show()
```

```python
plt.figure(figsize=[12,6], dpi=300)
sns.lineplot(x=range(log_history.shape[0]),y=log_history.accuracy,label='accuracy', color='#590242')
sns.lineplot(x=range(log_history_2i.shape[0],log_history_2i.shape[0]*2),y=log_history_2i.accuracy,label='loss 2nd iter', color='#590242', linestyle='--')
sns.lineplot(x=range(log_history.shape[0]),y=log_history.val_accuracy,label='val accuracy', color='#F2522E')
sns.lineplot(x=range(log_history_2i.shape[0],log_history_2i.shape[0]*2),y=log_history_2i.val_accuracy,label='val loss 2nd iter', color='#F2522E', linestyle='--')
plt.show()
```

```python
plt.figure(figsize=[12,6], dpi=300)
sns.lineplot(x=range(log_history.shape[0]),y=np.log(log_history.lr),label='learning rate')
plt.show()
```
